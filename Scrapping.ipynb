```python
# import library
import csv
import json
from bs4 import BeautifulSoup
import requests
# Request to website and download HTML contents
url='https://www.instructables.com/Building-a-Self-Driving-Boat-ArduPilot-Rover/'
req=requests.get(url)
content=req.text
```

```python
csv_file = open('scrape.csv', 'w')

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline', 'video_link','subheading','summary','photo_link'])
```

```python
print(content)
```

```python
soup=BeautifulSoup(content)
```

```python
print(soup.prettify())
```

```python
article = soup.find('article')
```

```python
print(article.prettify())
```

```python
headline = article.h1.text
print(headline)
```

```python
summary = article.find('div', class_='step-body').p.text
print(summary)
```

```python
try:
    vid_src = article.find('iframe')['src']
    vid_id = vid_src.split('/')[4]
    vid_id = vid_id.split('?')[0]

    yt_link = f'https://youtube.com/watch?v={vid_id}'
except Exception as e:
    yt_link = None

print(yt_link)
```

```python
headline = article.h1.text
print(headline)
try:
    vid_src = article.find('iframe',class- = "videoset-item")['src']
    vid_id = vid_src.split('/')[4]
    vid_id = vid_id.split('?')[0]

    yt_link = f'https://youtube.com/watch?v={vid_id}'
except Exception as e:
    yt_link = None

print(yt_link)
```

```python
csv_writer.writerow([headline,yt_link])
```

```python
subheads = []
summarys = []
photo = []
for articles in soup.find_all('article'):
    subheading = articles.find_all('strong')
    for s in subheading:
        print(s.text)
        subheads.append(s.text)
print("------------------------------------------------------------------")
print("------------------------------------------------------------------")
print("------------------------------------------------------------------")
summary = article.find_all('p')
for i in summary:
    print(i.text)
    summarys.append(i.text)
print("-------------------------------------------------------------------")
print("-------------------------------------------------------------------")
print("-------------------------------------------------------------------")
for image in soup.find_all('article'):
    img = image.find('img')['src']
    print(img)
    photo.append(img)
```

```python
subheads
```

```python
csv_writer.writerow([headline,yt_link,subheads])
```

```python
csvfile = open('scrape.csv', 'r')
jsonfile = open('file.json', 'w')

fieldnames = ('headline', 'summary','subheading')
reader = csv.DictReader( csvfile, fieldnames)
for row in reader:
    json.dump(row, jsonfile)
    jsonfile.write('\n')
```
